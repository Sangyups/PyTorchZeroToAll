##Optimizer 의 종류
- SGD
  - gradient 값을 0에 가깝게 만들어 로컬 미니멈을 구한다.
- Adagrad
  - step별로 lr값을 줄여나간다.
- Adam
  - sgd, momentum, RMSprop을 한꺼번에 사용(보통 권장됨)
- Adamax
  - infinity norm을 바탕으로 한 Adam의 다른 버전값을 0에 가깝게 만들어 로컬 미니멈을 구한다.
- Adagrad
   - step별로 lr값을 줄여나간다.
- Adam
   - sgd, momentum, RMSprop을 한꺼번에 사용(보통 권장됨)
- Adamax
   - infinity norm을 바탕으로 한 Adam의 다른 버전
- ASGD
  - SGD에서의 weight vector들의 평균을 취함
- LBFGS
- RMSprop
  - Adagrad에서 일정한 비율로 step 조절, gradient 0 수렴 문제 해결
- Rprop


```python

import torch
from torch.autograd import Variable

x_data = Variable(torch.Tensor([[1.0], [2.0], [3.0]]))
y_data = Variable(torch.Tensor([[2.0], [4.0], [6.0]]))

class Model(torch.nn.Module):
  def __init__(self):
    super(Model, self).__init__()
    self.linear = torch.nn.Linear(1, 1)
  def forward(self, x):
    y_pred = self.linear(x)
    return y_pred

model = Model()

criterion = torch.nn.MSELoss(size_average=False)
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

for epoch in range(500):
  y_pred =  model(x_data)

  loss = criterion(y_pred, y_data)
  print(epoch, loss.item())

  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

hour_var = Variable(torch.Tensor([4.0]))
print("predict (after training)", 4, model.forward(hour_var).item())
```
