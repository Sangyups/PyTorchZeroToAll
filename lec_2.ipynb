```python
import numpy as np
import matplotlib.pyplot as plt

x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

w = 1.0

def forward(x):
  return w * x #y_hat
def loss(x, y):
  y_pred = forward(x)
  return (y_pred - y) ** 2
def gradient(x,y):
  return 2 * x * (x * w - y)

print('predict (before training)', 4, forward(4))

for epoch in range(100):
  for x_val, y_val in zip(x_data, y_data):
    grad = gradient(x_val, y_val)
    w = w - 0.01 * grad
    print("\tgrad: ", x_val, y_val, grad)
    l = loss(x_val, y_val)
  print('progress:', epoch, 'w=', w, 'loss=', l)
print('predict (after training)', '4 hours', forward(4))
```

```python
#exercise3
#what if y = x^2 * 2 + x * 4
import numpy as np
import matplotlib.pyplot as plt

x_data = [1.0, 2.0, 3.0]
y_data = [6.0, 16.0, 30.0]

w1 = 1.0
w2 = 1.0

def forward(x):
  return x * x * w2 + x * w1 #y_hat
def loss(x, y):
  y_pred = forward(x)
  return (y_pred - y) ** 2
def gradient1(x,y):
  return 2 * x * (w2 * (x**2) + w1 * x - y)
def gradient2(x,y):
  return 2 * (x**2) * (w2 * (x**2) + w1 * x - y)

print('predict (before training)', 4, forward(4))

for epoch in range(200):
  for x_val, y_val in zip(x_data, y_data):
    grad1 = gradient1(x_val, y_val)
    grad2 = gradient2(x_val, y_val)
    w1 = w1 - 0.02 * grad1
    w2 = w2 - 0.01 * grad2
    print("\tgrad: ", x_val, y_val, grad1, grad2)
    l = loss(x_val, y_val)
  print('progress:', epoch, 'w1=', w1, 'w2=', w2, 'loss=', l)
print('predict (after training)', '4 hours', forward(4))
```
